Let's check if other types of phase transition occours: Graph of M(T) for a couple of cases

```python
import networkx as nx
from isingNetwork import IsingModel
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
t_range = np.linspace(1, 30, 100)
h_range = [0.01, 0.1, 1, 10, 100]
h_strings = [r'$10^{-2}$', r'$10^{-1}$', r'$1$', r'$10^1$', r'$10^2$']
```

Case 6. Different size, different connections

```python
sizes = [30, 470]  # Block sizes
p = [[0.8, 0.05],
     [0.05, 0.2]]  # Block connectivity probabilities
g = nx.stochastic_block_model(sizes, p)

adj_matrix = nx.adjacency_matrix(g)
plt.imshow(adj_matrix.toarray(), cmap='hot_r')

#plt.gca().invert_yaxis()

plt.xlabel('site i')
plt.ylabel('site j')
plt.show()
```

```python
model = IsingModel(g, temperature_range=t_range, iterations=20000, simmetric=False)

M = []
for h in tqdm(h_range):
    model.set_external_field(h)
    model.iterate()
    M.append(model.get_data('magnetization_per_spin'))

#save data
np.savez('data\\SBM_6_M(T)_N=500_iteration=20000_J=1.npz', *M)

#show data
for i in range(len(M)):
    plt.plot(t_range, M[i], label='h='+ h_strings[i])

plt.legend()
plt.xlabel(r'Temperature, $T$')
plt.ylabel(r'Order parameter, $M$')
plt.show()
```
